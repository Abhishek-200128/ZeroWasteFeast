```python
import pandas as pd

# Load the CSV file
file_path = 'waste.csv'
data = pd.read_csv(file_path)

# Get unique values for all columns
unique_values = {column: data[column].unique() for column in data.columns}

# Print unique values for each column
for column, values in unique_values.items():
    print(f"Unique values in '{column}':")
    print(values)
    print("\n")
```

```python
filtered_data = data[data['Jurisdiction'] != 'Tas']
```

```python
filtered_data = filtered_data[filtered_data['Type'] == 'Food organics']
```

```python
filtered_data
```

```python
filtered_data['Tonnes'] = filtered_data['Tonnes'].str.replace(',', '').astype(float)
```

```python
filtered_data = filtered_data[filtered_data['Tonnes'] != 0]
```

```python
filtered_data
```

```python
len(filtered_data)
```

```python
import matplotlib.pyplot as plt

# Filter out rows with missing or non-numeric values in 'Tonnes' column

filtered_data['Tonnes'] = pd.to_numeric(filtered_data['Tonnes'], errors='coerce')

# Group by 'Sub-stream' and calculate the sum of 'Tonnes'
tonnes_by_substream = filtered_data.groupby('Sub-stream')['Tonnes'].sum().sort_values(ascending=False)

# Plotting the data
plt.figure(figsize=(10, 6))
tonnes_by_substream.plot(kind='bar')
plt.title('Sum of Tonnes in Each Sub-stream')
plt.xlabel('Sub-stream')
plt.ylabel('Sum of Tonnes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
